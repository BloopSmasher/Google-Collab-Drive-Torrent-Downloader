**Download Torrent through Mangent Link**

```python
MAGNET_LINK = input("Enter your magnet link: ").strip()
MAGNET_LINK

!aria2c --dir="$DOWNLOAD_DIR" \
       --max-connection-per-server=4 \
       --split=4 \
       --min-split-size=1M \
       --auto-file-renaming=false \
       --console-log-level=warn \
       --seed-time=0 \
       --seed-ratio=0 \
       "$MAGNET_LINK" \
       2>&1 | tee /content/aria2.log
# Note - There is no Seeding

```
